<h1>Cleaning</h1>

```python
df.drop(["Date","Address","Postcode","YearBuilt","Lattitude","Longtitude"],axis=1,inplace=True)
```

```python
df.head()
```

#### Checking For Null Values

```python
df.isnull().sum()
```

#### Filling Null Values

```python
for i in ("Propertycount","Distance","Bedroom2"
         ,"Bathroom","Car"):
    df[i]=df[i].fillna(0)
```

```python
for i in ("Landsize","BuildingArea"):
    df[i]=df[i].fillna(df[i].mean())
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

## Encoding Features

### OneHotEncoding

```python
df
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

#### Defining Objects For Features & Target

```python
features=df.drop("Price",axis=1)
```

```python
features.columns
```

```python
target=df["Price"]
```

```python
features_train, features_test, target_train, target_test = train_test_split(features,target, test_size=0.2,random_state=12)
```

## Model Selection

### Linear Regression & OneHotEncoding

```python
model=LinearRegression()
```

```python
model.fit(features_train,target_train)
```

```python
predicted_price=model.predict(features_test)
predicted_price
```

### Checking For Over Fitting or Under Fitting

```python
model.score(features_train,target_train)
```

```python
model.score(features_test,target_test)
```

```python
print('Mean Squared Error ',mean_squared_error(target_test,predicted_price))
```

```python
print('Root Mean Squared Error ',np.sqrt(mean_squared_error(target_test,predicted_price)))
```

```python
print("Mean Absolute Error ",mean_absolute_error(target_test,predicted_price))
```

## Applying All Algorithm

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor,AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
models={'lr':LinearRegression(),
        'ls':Lasso(),
        'rg':Ridge(),
        'tree':DecisionTreeRegressor(),
        'rf':RandomForestRegressor(),
        'gb':GradientBoostingRegressor(),
        'knn':KNeighborsRegressor(),
        'svm':SVR(),
        'ad':AdaBoostRegressor(),
        'xbg':XGBRegressor()
       }
```

```python
for name, model in models.items():
    model1=model.fit(features_train,target_train)
    y_pred = model1.predict(features_test)
    print(f'{name} Mean squared error:  {mean_squared_error(target_test,y_pred)}') 
    print(f'{name} accuracy score: {model1.score(features_test,target_test)}\n')
```
